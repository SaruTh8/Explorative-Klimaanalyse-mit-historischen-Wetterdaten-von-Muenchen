```python
import pandas as pd
import matplotlib.pyplot as plt

#Monate und Jahr definieren
jahre = [str(jahr) for jahr in range(2005, 2026)]
monate = [str(monat).zfill(2) for monat in range(1, 13)]

#Basis-URL
base_url = "https://www.en.meteo.physik.uni-muenchen.de/~quicklooks/klimadaten/stadt/{jahr}/{monat}{jahr}_h.html"

#Liste für DataFrames
alle_df =  []

spalten = ["Tag", "T_max", "T_min", "Regen", "p_max", "p_min", "Sonne"]

#Schleife, die die Daten, die ich benötige extrahiert
for jahr in jahre:
    for monat in monate:
        url = base_url.format(jahr=jahr, monat=monat)
        try:
            df = pd.read_html(url)[0]
            df = df[spalten]
            df["Monat"] = monat #Spalte Monat wird hinzugefügt
            alle_df.append(df)
            print(f"{monat}.{jahr} geladen.")
        except Exception as e:
            print(f"Fehler bei {monat}: {e}")
        
#Alle DataFrames zusammenfügen
df_gesamt = pd.concat(alle_df, ignore_index=True)

#Überprüfung, ob der Merge der Daten funktioniert hat
#print(df_gesamt.head())
#print(df_gesamt.shape)
```

```python
print(df_gesamt.head())
```

```python
print(df_gesamt.shape)
print(df_gesamt.dtypes)
df_gesamt.head(10)
```

```python
# Datum parsen
df_gesamt["Datum"] = pd.to_datetime(df_gesamt["Tag"], format="%d.%m.%Y", errors="coerce")
df_gesamt = df_gesamt.drop(columns=["Tag", "Monat"])
df_gesamt = df_gesamt.set_index("Datum").sort_index()

# Numerische Spalten erzwingen
num_cols = ["T_max", "T_min", "p_max", "p_min"]
for col in num_cols:
    df_gesamt[col] = pd.to_numeric(df_gesamt[col], errors="coerce")

# Sanity-Check
print(df_gesamt.dtypes)
print(f"\nFehlende Werte:\n{df_gesamt.isna().sum()}")
print(f"\nZeitraum: {df_gesamt.index.min()} → {df_gesamt.index.max()}")
print(df_gesamt.describe().round(2))
```

```python
# Wo liegen die NaN-Lücken?
print("=== T_max NaN-Tage ===")
print(df_gesamt[df_gesamt["T_max"].isna()].index.tolist())

# Sind es dieselben Tage wie T_min?
gleiche_luecken = (df_gesamt["T_max"].isna() == df_gesamt["T_min"].isna()).all()
print(f"\nT_max und T_min NaN an denselben Tagen: {gleiche_luecken}")

# Sonne: Tages- oder Monatswerte?
print(f"\nSonne-Werte über 16h:")
print(df_gesamt[df_gesamt["Sonne"] > 16][["Sonne"]].head(10))
```

```python
# NaT-Zeilen entfernen
df_gesamt = df_gesamt[df_gesamt.index.notna()]
print(f"Shape nach NaT-Drop: {df_gesamt.shape}")
print(f"Verbleibende NaN:\n{df_gesamt.isna().sum()}")

# Sonne: laufende Monatssumme → Tageswert zurückrechnen
# Differenz zum Vortag, aber nur innerhalb desselben Monats
df_gesamt["Sonne_tag"] = df_gesamt.groupby(
    [df_gesamt.index.year, df_gesamt.index.month]
)["Sonne"].diff().clip(lower=0)

# Erster Tag des Monats hat keinen Vortag → Originalwert nehmen
erste_tage = df_gesamt.groupby(
    [df_gesamt.index.year, df_gesamt.index.month]
).head(1).index
df_gesamt.loc[erste_tage, "Sonne_tag"] = df_gesamt.loc[erste_tage, "Sonne"]

# Plausibilitätscheck
print(f"\nSonne_tag max: {df_gesamt['Sonne_tag'].max():.1f}h")
print(f"Sonne_tag Werte über 16h: {(df_gesamt['Sonne_tag'] > 16).sum()}")
```

```python
# Welche 3 Tage sind das?
print(df_gesamt[df_gesamt["Sonne_tag"] > 16][["Sonne", "Sonne_tag"]])
```

```python
# Sonne-Rohdaten durch Tageswerte ersetzen, Original-Sonne droppen
df_gesamt["Sonne"] = df_gesamt["Sonne_tag"]
df_gesamt = df_gesamt.drop(columns=["Sonne_tag"])

print(df_gesamt.describe().round(2))
```

```python
ausreisser = df_gesamt[df_gesamt["Sonne"] > 16].index
df_gesamt.loc[ausreisser, "Sonne"] = float("nan")
df_gesamt["Sonne"] = df_gesamt["Sonne"].interpolate(method="time")

print(f"Sonne max nach Fix: {df_gesamt['Sonne'].max():.1f}h")
print(f"Verbleibende NaN: {df_gesamt['Sonne'].isna().sum()}")
```
